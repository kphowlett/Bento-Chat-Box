```python
def upload_df(file):
    df = pd.read_csv(file, encoding="UTF-8", index_col=False)
    df = df.rename(columns={
    "demographic_data_id": "id",
    "age_at_index": "age",
    }).drop(columns = ["type", "study_subject.study_subject_id"]).reset_index(drop=True)
    return df
```

```python
upload_df('upload_file4.csv')
```
